### Exercise 1
Calculate the mean, median and mode for this data set: 
```
19, 18, 21, 16, 15, 17, 20, 18
```
While comparing the results of three measures, comment about this distribution. 

````python
#Mean 19+18+21+16+15+17+20+18= 144; 144/8(number of values) = 18
#Median 15,16,17,18,18,19,20,21; there are an even number of values so we average the two centermostvalues ((18+18)/2)= 18
#Mode 18 appears twice, all others once
#Mean = 18    Median= 18     Mode = 18
#This data set is symmetrical, all of its measures of central tendancy align at 18
````

### Exercise 3
On the first three days of his bookshop opening, Joe sold 15, 18, and 16 books (He initially hoped that he would sell 17 books every day).  How many books does he need to sell on the next day to have a mean sale of 17?

````python
#(15+18+16+x)/4= 17
#15+18+16+x= 68
#x= 68-15-18-16
#x= 19
# He needs to sell 19 books on the fourth day to bring his average up to 17
````
